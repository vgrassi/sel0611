```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros do modelo do pêndulo sem amortecimento:
m = 1
g = 9.81
l = 1

# Função de transferência do pêndulo sem amortecimento:
num = np.array([1/(m*(l**2))])
den = np.array([1, 0, g/l])
H = ct.tf(num, den)

print('H(s) = {}'.format(H))
```

Resposta no tempo:

```python
# Resposta ao degrau unitário:
(t, theta) = ct.step_response(H)
plt.plot(t, theta)
plt.xlabel('Time (s)')
plt.ylabel('Angle (rad)')
plt.title('Step response')
plt.show()
```
