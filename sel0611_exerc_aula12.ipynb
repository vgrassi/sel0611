## Modelo de pêndulo sem amortecimento.

```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros do modelo do pêndulo sem amortecimento:
m = 1
g = 9.81
l = 1

# Função de transferência do pêndulo sem amortecimento:
num = np.array([1/(m*(l**2))])
den = np.array([1, 0, g/l])
H = ct.tf(num, den)

print('H(s) = {}'.format(H))
```

Resposta no tempo:

```python
# Resposta ao degrau unitário:
(t, theta) = ct.step_response(H)
plt.plot(t, theta)
plt.xlabel('Time (s)')
plt.ylabel('Angle (rad)')
plt.title('Step response')
plt.show()
```

Pólos e zeros do sistema:

```python
# Pólos and zeros:
(p, z) = ct.pzmap(H)
p = np.reshape(p, (len(p), 1))
print('Poles of H(s): \n{}'.format(p))
```

Diagrama de bode:

```python
# Diagrama de Bode:
#plt.figure(2)
ct.bode_plot(H, dB=True)
plt.show(block=False)
```
